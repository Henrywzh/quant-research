```python
import re
import numpy as np

def parse_aum(x):
    """
    Parse strings like '92.43M', '4.79B', '139.23M' into RMB numeric.
    Assumes:
      M = million, B = billion
    """
    if pd.isna(x):
        return np.nan
    s = str(x).strip()
    if s in ["-", "—", "nan", "None", ""]:
        return np.nan

    m = re.fullmatch(r"([0-9]*\.?[0-9]+)\s*([KMB])", s, flags=re.IGNORECASE)
    if m:
        val = float(m.group(1))
        unit = m.group(2).upper()
        mult = {"K": 1e3, "M": 1e6, "B": 1e9}[unit]
        return val * mult

    # fallback: try numeric
    try:
        return float(s.replace(",", ""))
    except:
        return np.nan

etf_all['资产管理规模(¥)'] = etf_all['资产管理规模(¥)'].apply(parse_aum)
```

```python
"""
A股行业/主题 ETF Universe + OHLCV 拉取 + 标准化 + 质量检查（Step 1 -> 5）
- 数据源：AkShare 东方财富（ETF/LOF）
- OHLCV 起始日期：20150101
- 输出：
  1) meta_universe.csv：universe 元数据（含 instrument_type）
  2) ohlcv_long.parquet：长表行情（date,ticker,open,high,low,close,volume,amount,pct_chg,turnover）
  3) ohlcv_wide_{field}.parquet：宽表（dates x tickers）
  4) universe_quality_report.csv：质量报告（长度、缺失、最新日期、60日成交额等）
"""

from __future__ import annotations

import os
import time
from dataclasses import dataclass
from datetime import datetime
from typing import Dict, Iterable, List, Optional, Tuple

import numpy as np
import pandas as pd

try:
    import akshare as ak
except ImportError as e:
    raise ImportError("Please install akshare first: pip install akshare") from e


# -----------------------------
# Step 1: Universe 分层（行业池 vs 主题池）
# -----------------------------
def build_universe_meta(
    sw_sector_etf: Dict[str, str],
    other_sector_etf: Dict[str, str],
) -> pd.DataFrame:
    """
    返回 universe meta（不含 instrument_type），并区分 group_type:
      - group_type='industry'  -> sw_sector_etf
      - group_type='theme'     -> other_sector_etf
    """
    rows = []
    for k, v in sw_sector_etf.items():
        rows.append({"group_type": "industry", "group_name": k, "code": str(v)})
    for k, v in other_sector_etf.items():
        rows.append({"group_type": "theme", "group_name": k, "code": str(v)})

    df = pd.DataFrame(rows)
    df["code"] = df["code"].astype(str).str.strip()
    # 去重：同一代码可能被你放在不同桶里（保留第一条，后续你可自行处理）
    df = df.drop_duplicates(subset=["code"], keep="first").reset_index(drop=True)
    return df


# -----------------------------
# Step 2: 标的类型识别（ETF vs LOF）
# -----------------------------
def infer_instrument_type_by_spot(codes: Iterable[str]) -> pd.DataFrame:
    """
    用 akshare 的 spot 列表来判断 instrument_type:
      - ETF: code in ak.fund_etf_spot_em()
      - LOF: code in ak.fund_lof_spot_em()
      - UNKNOWN: neither
    返回：DataFrame(code, instrument_type)
    """
    codes = [str(c).strip() for c in codes]

    # ETF spot
    etf_spot = ak.fund_etf_spot_em()
    # 常见列：['代码', '名称', ...]；做稳健处理
    etf_code_col = "代码" if "代码" in etf_spot.columns else etf_spot.columns[0]
    etf_codes = set(etf_spot[etf_code_col].astype(str).str.strip().tolist())

    # LOF spot
    lof_spot = ak.fund_lof_spot_em()
    lof_code_col = "代码" if "代码" in lof_spot.columns else lof_spot.columns[0]
    lof_codes = set(lof_spot[lof_code_col].astype(str).str.strip().tolist())

    out = []
    for c in codes:
        if c in etf_codes:
            t = "ETF"
        elif c in lof_codes:
            t = "LOF"
        else:
            t = "UNKNOWN"
        out.append({"code": c, "instrument_type": t})

    return pd.DataFrame(out)


# -----------------------------
# Step 3: 拉取行情（ETF/LOF）并统一复权口径
# -----------------------------
CN2EN_COLS = {
    "日期": "date",
    "开盘": "open",
    "收盘": "close",
    "最高": "high",
    "最低": "low",
    "成交量": "volume",
    "成交额": "amount",
    "涨跌幅": "pct_chg",
    "换手率": "turnover",
    # 可选列，不强制使用
    "振幅": "amplitude",
    "涨跌额": "chg_amt",
}


def _standardize_hist_df(df: pd.DataFrame, ticker: str) -> pd.DataFrame:
    """
    将 akshare 返回的东方财富 ETF/LOF 历史行情标准化为固定 schema。
    输出列：date,ticker,open,high,low,close,volume,amount,pct_chg,turnover
    """
    if df is None or df.empty:
        return pd.DataFrame(
            columns=[
                "date",
                "ticker",
                "open",
                "high",
                "low",
                "close",
                "volume",
                "amount",
                "pct_chg",
                "turnover",
            ]
        )

    # 统一列名
    rename = {c: CN2EN_COLS[c] for c in df.columns if c in CN2EN_COLS}
    df = df.rename(columns=rename).copy()

    # 兼容：有些返回可能已经是英文或混合
    if "date" not in df.columns and "日期" in df.columns:
        df["date"] = df["日期"]
    df["date"] = pd.to_datetime(df["date"], errors="coerce")

    df["ticker"] = str(ticker)

    # 保留字段
    keep = ["date", "ticker", "open", "high", "low", "close", "volume", "amount", "pct_chg", "turnover"]
    for col in keep:
        if col not in df.columns:
            df[col] = np.nan

    df = df[keep].copy()

    # 数值化
    num_cols = ["open", "high", "low", "close", "volume", "amount", "pct_chg", "turnover"]
    for c in num_cols:
        df[c] = pd.to_numeric(df[c], errors="coerce")

    # 排序、去重
    df = df.dropna(subset=["date"]).sort_values("date")
    df = df.drop_duplicates(subset=["date", "ticker"], keep="last").reset_index(drop=True)

    return df


def fetch_hist_em(
    symbol: str,
    instrument_type: str,
    start_date: str = "20150101",
    end_date: Optional[str] = None,
    adjust: str = "hfq",
    period: str = "daily",
    sleep_s: float = 0.2,
) -> pd.DataFrame:
    """
    拉取单只标的历史行情（东财），并返回标准化长表。
    instrument_type: 'ETF' or 'LOF'
    adjust: ''(不复权) / 'qfq'(前复权) / 'hfq'(后复权)
    """
    if end_date is None:
        end_date = datetime.today().strftime("%Y%m%d")

    symbol = str(symbol).strip()

    try:
        if instrument_type == "ETF":
            raw = ak.fund_etf_hist_em(symbol=symbol, period=period, start_date=start_date, end_date=end_date, adjust=adjust)
        elif instrument_type == "LOF":
            raw = ak.fund_lof_hist_em(symbol=symbol, period=period, start_date=start_date, end_date=end_date, adjust=adjust)
        else:
            # UNKNOWN：直接返回空
            raw = pd.DataFrame()
    except Exception as e:
        print(f"[WARN] fetch failed: symbol={symbol}, type={instrument_type}, err={repr(e)}")
        raw = pd.DataFrame()

    time.sleep(max(0.0, sleep_s))
    return _standardize_hist_df(raw, ticker=symbol)


def fetch_universe_ohlcv(
    meta: pd.DataFrame,
    start_date: str = "20150101",
    end_date: Optional[str] = None,
    adjust: str = "hfq",
    period: str = "daily",
    sleep_s: float = 0.2,
) -> pd.DataFrame:
    """
    批量拉取 universe 行情，返回 long_df。
    meta 需包含：code, instrument_type
    """
    assert "code" in meta.columns
    assert "instrument_type" in meta.columns

    frames = []
    for _, row in meta.iterrows():
        code = str(row["code"]).strip()
        itype = str(row["instrument_type"]).strip()
        df = fetch_hist_em(
            symbol=code,
            instrument_type=itype,
            start_date=start_date,
            end_date=end_date,
            adjust=adjust,
            period=period,
            sleep_s=sleep_s,
        )
        if not df.empty:
            frames.append(df)

    if not frames:
        return pd.DataFrame(
            columns=["date", "ticker", "open", "high", "low", "close", "volume", "amount", "pct_chg", "turnover"]
        )

    long_df = pd.concat(frames, ignore_index=True)
    long_df = long_df.sort_values(["date", "ticker"]).reset_index(drop=True)
    return long_df


# -----------------------------
# Step 4: 数据结构化（长表 -> 宽表面板）
# -----------------------------
def build_wide_panels(long_df: pd.DataFrame) -> Dict[str, pd.DataFrame]:
    """
    输出 dict[field] = DataFrame(index=date, columns=ticker)
    fields: open, high, low, close, volume, amount, pct_chg, turnover
    """
    fields = ["open", "high", "low", "close", "volume", "amount", "pct_chg", "turnover"]
    panels = {}
    for f in fields:
        wide = long_df.pivot(index="date", columns="ticker", values=f).sort_index()
        panels[f] = wide
    return panels


# -----------------------------
# Step 5: Universe 质量检查报告
# -----------------------------
def universe_quality_report(
    long_df: pd.DataFrame,
    meta: pd.DataFrame,
    extreme_pct_chg: float = 0.20,
    last_n_liq: int = 60,
) -> pd.DataFrame:
    """
    质量检查指标：
      - n_days: 有效交易日数
      - start_date / end_date
      - missing_rate_close: 在 union calendar 上 close 缺失率
      - last_date_gap_days: 距离全体最大日期的滞后天数
      - amount_mean_{N}: 近N日成交额均值
      - n_nonpos_close: close<=0 的天数
      - n_extreme_pct: |pct_chg| > extreme_pct_chg 的天数（提示异常）
    """
    meta0 = meta.copy()
    meta0["code"] = meta0["code"].astype(str).str.strip()

    # calendar 对齐（用 close 宽表）
    close_wide = long_df.pivot(index="date", columns="ticker", values="close").sort_index()
    max_date = close_wide.index.max() if len(close_wide.index) else pd.NaT

    rep_rows = []
    for code in meta0["code"].tolist():
        t = str(code).strip()
        sub = long_df[long_df["ticker"] == t].sort_values("date")

        if sub.empty:
            rep_rows.append(
                {
                    "code": t,
                    "instrument_type": meta0.loc[meta0["code"] == t, "instrument_type"].values[0]
                    if "instrument_type" in meta0.columns and (meta0["code"] == t).any()
                    else "UNKNOWN",
                    "n_days": 0,
                    "start_date": pd.NaT,
                    "end_date": pd.NaT,
                    "missing_rate_close": 1.0,
                    "last_date_gap_days": np.nan,
                    f"amount_mean_{last_n_liq}": np.nan,
                    "n_nonpos_close": np.nan,
                    "n_extreme_pct": np.nan,
                }
            )
            continue

        start = sub["date"].min()
        end = sub["date"].max()
        n_days = sub["date"].nunique()

        # 缺失率（在 union calendar 上）
        if t in close_wide.columns:
            missing_rate = float(close_wide[t].isna().mean())
        else:
            missing_rate = 1.0

        # 最新滞后
        gap = (max_date - end).days if pd.notna(max_date) and pd.notna(end) else np.nan

        # 近 N 日成交额均值（按日期最新 N 行）
        amount_mean = float(sub.tail(last_n_liq)["amount"].mean())

        # 异常
        n_nonpos = int((sub["close"] <= 0).sum())
        n_extreme = int((sub["pct_chg"].abs() > extreme_pct_chg).sum())

        rep_rows.append(
            {
                "code": t,
                "instrument_type": meta0.loc[meta0["code"] == t, "instrument_type"].values[0]
                if "instrument_type" in meta0.columns and (meta0["code"] == t).any()
                else "UNKNOWN",
                "n_days": n_days,
                "start_date": start,
                "end_date": end,
                "missing_rate_close": missing_rate,
                "last_date_gap_days": gap,
                f"amount_mean_{last_n_liq}": amount_mean,
                "n_nonpos_close": n_nonpos,
                "n_extreme_pct": n_extreme,
            }
        )

    rep = pd.DataFrame(rep_rows)

    # 合并 group 信息（行业/主题标签）
    if "group_type" in meta0.columns and "group_name" in meta0.columns:
        rep = rep.merge(meta0[["code", "group_type", "group_name"]], on="code", how="left")

    # 排序：先看行业池，再看主题池；再按成交额/历史长度
    sort_cols = []
    if "group_type" in rep.columns:
        sort_cols.append("group_type")
    sort_cols += [f"amount_mean_{last_n_liq}", "n_days"]
    rep = rep.sort_values(sort_cols, ascending=[True] + [False] * (len(sort_cols) - 1)).reset_index(drop=True)

    return rep


# -----------------------------
# 一键运行（Step 1 -> 5）
# -----------------------------
@dataclass
class RunConfig:
    start_date: str = "20150101"
    end_date: Optional[str] = None
    adjust: str = "hfq"
    period: str = "daily"
    sleep_s: float = 0.2
    out_dir: str = "data_cn_etf_universe"
    extreme_pct_chg: float = 0.20
    last_n_liq: int = 60


def run_step_1_to_5(
    sw_sector_etf: Dict[str, str],
    other_sector_etf: Dict[str, str],
    cfg: RunConfig = RunConfig(),
) -> Tuple[pd.DataFrame, pd.DataFrame, Dict[str, pd.DataFrame], pd.DataFrame]:
    """
    返回：
      - meta（含 instrument_type）
      - long_df
      - panels（宽表 dict）
      - report（质量报告）
    并落盘保存到 cfg.out_dir。
    """
    os.makedirs(cfg.out_dir, exist_ok=True)

    # Step 1
    meta = build_universe_meta(sw_sector_etf, other_sector_etf)

    # Step 2
    types = infer_instrument_type_by_spot(meta["code"].tolist())
    meta = meta.merge(types, on="code", how="left")

    # 先剔除 UNKNOWN（你也可以选择保留并另行处理）
    meta_valid = meta[meta["instrument_type"].isin(["ETF", "LOF"])].reset_index(drop=True)

    # 保存 meta
    meta.to_csv(os.path.join(cfg.out_dir, "meta_universe.csv"), index=False, encoding="utf-8-sig")
    meta_valid.to_csv(os.path.join(cfg.out_dir, "meta_universe_valid.csv"), index=False, encoding="utf-8-sig")

    # Step 3
    long_df = fetch_universe_ohlcv(
        meta_valid,
        start_date=cfg.start_date,
        end_date=cfg.end_date,
        adjust=cfg.adjust,
        period=cfg.period,
        sleep_s=cfg.sleep_s,
    )
    long_path = os.path.join(cfg.out_dir, "ohlcv_long.parquet")
    long_df.to_parquet(long_path, index=False)

    # Step 4
    panels = build_wide_panels(long_df)
    for k, df_wide in panels.items():
        df_wide.to_parquet(os.path.join(cfg.out_dir, f"ohlcv_wide_{k}.parquet"))

    # Step 5
    report = universe_quality_report(
        long_df,
        meta_valid,
        extreme_pct_chg=cfg.extreme_pct_chg,
        last_n_liq=cfg.last_n_liq,
    )
    report.to_csv(os.path.join(cfg.out_dir, "universe_quality_report.csv"), index=False, encoding="utf-8-sig")

    return meta, long_df, panels, report
```

```python
cfg = RunConfig(
        start_date="20150101",
        adjust="hfq",
        sleep_s=0.15,
        out_dir="data_cn_etf_universe",
    )

meta, long_df, panels, report = run_step_1_to_5(sw_sector_etf, other_sector_etf, cfg)

print("Saved to:", cfg.out_dir)
print("Meta valid:", meta[meta["instrument_type"].isin(["ETF", "LOF"])].shape)
print("Long df:", long_df.shape)
print("Report head:\n", report.head(10))
```
