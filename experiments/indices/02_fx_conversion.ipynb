## Bitcoin ETF vs Bitcoin

```python
btc_tickers = ['BTC-USD', 'IBIT']
btc_md = download_market_data(btc_tickers, start='2000-01-01')
btc_close_df = btc_md.close
btc_cleaned = btc_close_df.dropna()
btc_cleaned /= btc_cleaned.iloc[0]
btc_cleaned.plot()
```
